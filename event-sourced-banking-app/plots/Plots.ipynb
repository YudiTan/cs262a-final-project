```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Banking Application Plots

```python
# 1000 Events
df = pd.DataFrame({
    'Snapshot Interval': [100, 300, 500, 1000],
    'Python': [0.591, 0.589, 0.609, 0.634],
    'Ray': [1.18, 1.192, 1.198, 1.215]
})
fig, ax1 = plt.subplots(figsize=(5, 5))
tidy = df.melt(id_vars='Snapshot Interval').rename(columns=str.title)
ax = sns.barplot(x='Snapshot Interval', y='Value', hue='Variable', data=tidy, ax=ax1)
ax.set(xlabel='Snapshot Intervals', ylabel='Mean Latency (s)')
ax.set_title("1000 Events")
sns.despine(fig)
```
