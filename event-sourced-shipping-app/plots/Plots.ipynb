```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Shipping Application Plots

```python
# 5000 Events
df = pd.DataFrame({
    'Actors': [1, 2, 3, 4, 5, 20],
    'Python': [9.912, 10.11, 10.235, 9.89, 10.245, 9.981],
    'Ray': [13.318, 12.159, 9.961, 9.878, 9.625, 14.473]
})
fig, ax1 = plt.subplots(figsize=(5, 5))
tidy = df.melt(id_vars='Actors').rename(columns=str.title)
ax = sns.barplot(x='Actors', y='Value', hue='Variable', data=tidy, ax=ax1)
ax.set(xlabel='Number of Actors', ylabel='Mean Latency (s)')
ax.set_title("5000 Events")
sns.despine(fig)
```
